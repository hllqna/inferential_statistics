# Standard Deviation and Variance
The standard deviation of a population (normalized by N) is less than a standard deviation of a sample (normalized by N-1).

Variance is std squared.

[Back to top](#concepts-covered)

```python
# OOBP
print(f"OOBP mean: {df.OOBP.mean()}")

# Std and var of a sample (normalized by N-1):
print(f"OOBP std (sample): {df.OOBP.std()}")
print(f"OOBP var (sample): {df.OOBP.var()}")

# Std and var of a population (normalized by N):
print(f"OOBP std (population): {df.OOBP.std(ddof=0)}")
print(f"OOBP var (population): {df.OOBP.var(ddof=0)}")
```

```python
# std of sample of BA
print(f"STD (sample) BA: {df.BA.std()}")

# var (population) of BA
print(f"Variance (population) BA: {df.BA.var(ddof=0)}")

df.BA.hist()
```

## Coefficient of Variation (CV)
A measure of data variability expressed as a percentage of the mean. It is the ratio of standard deviation to mean, allowing standardized comparison of datasets.
[Back to top](#concepts-covered)

```python
# CV for Runs Scored
print(f"Coefficient of variation of RS: {df.RS.std()/df.RS.mean()}")

#As a rule of thumb, the data with a high variability has CV > 1, otherwise we could say that it has low variability
```

```python
# CV for Wins
print(df.W.std()/df.W.mean())
```

# Covariance

indicates the direction of the linear relationship between two variables. If covariance > 0, the linear relationship is positive; covariance < 0 indicates negative linear relationship; covariance = 0 meaning no linear relationship.

Covariance does not indicate the strength of the relationship

[Back to top](#concepts-covered)

```python
# Covariance works with continuous variables (and not so well with discrete ones)

df_co = df[["RS", "RA", "W"]]
df_co.cov()
```

```python
df_co2 = df[["OBP", "SLG", "OOBP", "OSLG"]]
df_co2.cov()
```

# Correlation

There are different types of correlation: Pearson, Kendall, Spearman, etc.. Pearson correlation is the most commonly used one. 

The correlation measures the strength of a relationship between 2 variables. It varies between -1 and 1. Corr = 1 indicates strong positive relation, -1 - strong negative relation, 0 - no relation.

Correlation is not causation (search for spurious correlations).

[Back to top](#concepts-covered)

```python
df_co.corr()

# Rule of Thumb:
# Correlation > 0.5 -> strong
# Correlation [0.3, 0.5) -> medium
# Correlation [0.1, 0.3) -> low
```

```python
# Visualization of correlation via heatmap
sns.heatmap(data = df_co.corr(),
            annot = True,
            center = 0,
            cmap = 'coolwarm',
            linewidth = 10)
```

```python
sns.heatmap(data = df_co2.corr(),
            annot = True,
            center = 0.5,
            cmap = 'twilight_shifted',
            linewidth = 1,
            linecolor='black',
            square = False,
            cbar = False)
```

# Normal Distribution

Normal (Gaussian) distribution has the rule:
68.26% of the observation are within 1 standard deviation from the mean,
95.44% are within 2 std from mean,
99.72% are within 3 std from mean.

So a statistics like "3% of customers purchased above 2 standard deviations" looks okay, whereas "3% of customers purchased above 3 standard deviations" is a dicrepancy.

6-sigma as a name of practices for continuous self-improvement comes from the goal of 6 sigma where 99.9966% of the products are defect-free.

[Back to top](#concepts-covered)

```python
# Plotting W
df.W.plot.density()
```

```python
# 68-95-99-check
print(f"% Within 1 std: {
df.loc[(df.W >= df.W.mean() - 1 * df.W.std()) &
       (df.W <= df.W.mean() + 1 * df.W.std())].W.count()*100/df.W.count()}")
print(f"% Within 2 std: {
df.loc[(df.W >= df.W.mean() - 2 * df.W.std()) &
       (df.W <= df.W.mean() + 2 * df.W.std())].W.count()*100/df.W.count()}")
print(f"% Within 3 std: {
df.loc[(df.W >= df.W.mean() - 3 * df.W.std()) &
       (df.W <= df.W.mean() + 3 * df.W.std())].W.count()*100/df.W.count()}")
```

```python
df.RA.plot.density()
```

```python
# 68-95-99-check for RA
for i in range(1,4):
    print(f"% Within {i} std: {
    df.loc[(df.RA >= df.RA.mean() - i * df.RA.std()) &
        (df.RA <= df.RA.mean() + i * df.RA.std())].RA.count()*100
        /df.RA.count()}")
```
