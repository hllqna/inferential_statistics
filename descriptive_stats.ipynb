```python
import pandas as pd
import seaborn as sns
import scipy.stats as st
```

```python
data_path = "Inf_Stats_files/Descriptive Statistics/Baseball.csv"
df = pd.read_csv(data_path)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Mean

```python
# Mean of Runs Scored
df.RS.mean()
```

```python
# Mean of Runs Scored by the Arizona Team (ARI)
df.loc[df.Team == 'ARI'].RS.mean()
```

```python
# Mean of Run Scored by the Arizona Team since 2005
df.loc[(df.Team == 'ARI') & (df.Year >= 2005)].RS.mean()
```

```python
# Mean of Runs Allowed by the Chicago Team (CHC) before 2000 and after 2005
df.loc[(df.Team == 'CHC') & ((df.Year < 2000)|(df.Year > 2005))].RA.mean()
```

# Median

```python
# Compare mean and median of Wins 
print(df.W.mean())
print(df.W.median())
# close values indicate more uniformly distributed data, values far apart indicate skewness
```

```python
# Median of Wins of the Baltimore team (BAL) until 2000
df.loc[(df.Team == 'BAL') & (df.Year <= 2000)].W.median()
```

# Mode

```python
# Mode, mean and median of On-Base Percentage
print(df.OBP.mode())
print(df.OBP.mean())
print(df.OBP.median())
df.OBP.hist()
```

```python
# Mode of OBP during 2010
df.loc[df.Year == 2010].OBP.mode()
```

```python
# Most Common Season Ranking for Boston (BOS)
df.loc[df.Team == 'BOS'].RankSeason.mode()
```
