# Standard Deviation and Variance
The standard deviation of a population (normalized by N) is less than a standard deviation of a sample (normalized by N-1).

Variance is std squared.

[Back to top](#concepts-covered)

```python
# OOBP
print(f"OOBP mean: {df.OOBP.mean()}")

# Std and var of a sample (normalized by N-1):
print(f"OOBP std (sample): {df.OOBP.std()}")
print(f"OOBP var (sample): {df.OOBP.var()}")

# Std and var of a population (normalized by N):
print(f"OOBP std (population): {df.OOBP.std(ddof=0)}")
print(f"OOBP var (population): {df.OOBP.var(ddof=0)}")
```

```python
# std of sample of BA
print(f"STD (sample) BA: {df.BA.std()}")

# var (population) of BA
print(f"Variance (population) BA: {df.BA.var(ddof=0)}")

df.BA.hist()
```

## Coefficient of Variation (CV)
A measure of data variability expressed as a percentage of the mean. It is the ratio of standard deviation to mean, allowing standardized comparison of datasets.
[Back to top](#concepts-covered)

```python
# CV for Runs Scored
print(f"Coefficient of variation of RS: {df.RS.std()/df.RS.mean()}")

#As a rule of thumb, the data with a high variability has CV > 1, otherwise we could say that it has low variability
```

```python
# CV for Wins
print(df.W.std()/df.W.mean())
```
