# Standard Error of the Sample Mean
It estimates how far the sample mean is likely to be from the population mean.

Standard Error is calculated as follows:
$$
SE = \frac{\sigma}{\sqrt{n}}
$$
where $\sigma$ is standard deviation of the sample, $n$ is number of observations.

[Back to top](#concepts-covered)


```python
# SE of price
# through formula:
print(df.Price.std() / math.sqrt(df.Price.count())) 
# with stats package:
print(st.sem(df.Price))
```

# Z-Score and Standartization

Z-score estimates how far away a data point is from its mean.
$$
z = \frac{x - \mu}{\sigma}
$$

[Back to top](#concepts-covered)


```python
# Z-score for delivery time
# using formula
df["Delivery Time standardized"] = (df["Delivery Time"] - df["Delivery Time"].mean())/df["Delivery Time"].std()
```

```python
# using sklearn
df["Delivery Time standardized2"] = preprocessing.scale(df["Delivery Time"])
```

# Confidence Level and Confidence Interval

<b>Confidence level</b> indicates how sure one can be that a statistical conclusion is correct. Expressed in $\%$.

A typical confidence level is set at 95$\%$ chance that the results are accurate.

Confidence level is used in hypothesis testing and is used to calculate confidence intervals.

High confidence level doesn't mean higher accuracy, just higher certainty.

<b>Confidence interval</b> is a range likely to contain the true population parameter with a certain level certainty.

Confidence interval width is determined by:
- Standard error of the sample mean
- Confidence level

[Back to top](#concepts-covered)

```python
cl = [0.005, 0.025, 0.05, 0.95, 0.975, 0.995]
for alpha in cl:
    print(f"The corresponding value in normal distribution for {alpha} is: {round(st.norm.ppf(alpha), 3)}")
```

<b>Confidence Intervals for Large Samples</b>

$$
CI = \bar{x} \pm z \frac{s}{\sqrt{n}}
$$

Sample size must be above 30 (ideally above 100) observations.

Practical advice: not to make conclusions on samples of 10 or less observations.

Z-value represents the desired confidence level (often 95%). Confidence interval: from 2.5 to 97.5 percentile. To determine CI, find z value corresponding to 2.5 or 97.5% instead of 95%

<i>Example.</i>

A survey shows that 60% of customers would buy the new product. The company wants to know the survey's accuracy. Confidence interval can estimate the range of possible true populaton percentages. For instance, 95% confidence interval of 55-65% would mean 95% confidence that the true percentage of customers falls within that range. This helps recogizing the level of uncertainty in the predictions and evaluate the potential risks.

```python
# Confidence interval for the Price mean
print(f"The mean is {df.Price.mean()}")
print(f"The 95-confidence interval for it: {st.norm.interval(confidence=0.95,
                                            loc = df.Price.mean(),
                                            scale = st.sem(df.Price))}")
```

```python
# function to compute the confidence interval for the mean
# for each numerical variable that has a sample size bigger than 30
def conf_interval_for_mean(X: pd.Series, confidence=0.95)->tuple[float, float]:
    if not isinstance(X, pd.Series):
        raise TypeError("X must be a pandas Series.")
    if not 0 < confidence < 1:
        raise ValueError("confidence must be between 0 and 1.")
    if X.count() < 30:
        raise ValueError(f"At least 30 non-null observations are required, got {X.count()}.")

    x = X.dropna()
    mean = x.mean()
    se = st.sem(x)

    ci_low, ci_high = st.norm.interval(confidence=confidence, loc=mean, scale=se)
    return (float(ci_low), float(ci_high))
```

```python
def calculate_confidence_intervals(df, confidence=0.95):
    intervals = {}

    # Iterate over each numerical column in the dataframe
    for column in df.select_dtypes(include=[np.number]).columns:
        data = df[column].dropna()
        # Check if the sample size is greater than 30
        if len(data) > 30:
            # Compute the confidence interval for the mean
            interval = st.norm.interval(confidence, loc=data.mean(), scale=st.sem(data))
            intervals[column] = interval
    return intervals


print(calculate_confidence_intervals(df))
```
